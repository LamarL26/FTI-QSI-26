Create variables to track the number of hours of worked (42) and the hourly rate (15). Regular pay covers the first 40 hours. Any hours beyond 40 are paid at 1.5x the rate. Calculate and print total pay.

```python
# create variables

# write a conditional
```

##### **Solution**

```python
hours_worked = 42
hourly_rate = 15

over_hourly_rate = hourly_rate * 1.5

if hours_worked <= 40:
  print(hours_worked * hourly_rate)
else:
  under_40_pay = 40 * hourly_rate
  over_40_pay = (hours_worked - 40) * over_hourly_rate
  total_pay = under_40_pay + over_40_pay
  print(total_pay)
```
